# Customer Segment Analysis

## Objective

The objective of this analysis is to understand the purchasing behavior of different customer segments.

This section answers the following business questions:

- Which customer segment generates the highest sales?
- Which customer segment contributes the highest profit?
- Which customer segment places the highest number of orders?
- Which customer segment has the highest average sales per transaction?

These insights help businesses identify their most valuable customers and optimize marketing strategies accordingly.

```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../data/cleaned/superstore_cleaned.csv")
```

Sales by Customer Segmentation


```python
Sales_By_Segment = df.groupby("Segment")["Sales"].sum().sort_values(ascending = False)
print(Sales_By_Segment)
```

```python
plt.figure(figsize=(8,5))

Sales_By_Segment.plot(kind="bar")

plt.title("Total Sales by Customer Segment")
plt.xlabel("Customer Segment")
plt.ylabel("Total Sales")
plt.xticks(rotation=0)

plt.show()
```

Profit by Customer Segmnetation

```python
profit_by_segment = df.groupby("Segment")["Profit"].sum().sort_values(ascending=False)
print(profit_by_segment)
```

```python
plt.figure(figsize=(8,5))
profit_by_segment.plot(kind="bar")
plt.title("Total Profit by Customer Segment")
plt.xlabel("Customer Segment")
plt.ylabel("Total Profit")
plt.xticks(rotation=0)
plt.show()
```

Average Sales per Transaction

```python
average_sales_segment = df.groupby("Segment")["Sales"].mean().sort_values(ascending=False)
print(average_sales_segment)
```

```python
plt.figure(figsize=(8,5))
average_sales_segment.plot(kind="bar")
plt.title("Average Sales per Transaction by Customer Segment")
plt.xlabel("Customer Segment")
plt.ylabel("Average Sales")
plt.xticks(rotation=0)
plt.show()
```

Quantity Sold By Segment

```python
quantity_by_segment =df.groupby("Segment")["Quantity"].sum().sort_values(ascending=False)
print(quantity_by_segment)
```

```python
plt.figure(figsize=(8,5))
quantity_by_segment.plot(kind="bar")
plt.title("Quantity Sold by Customer Segment")
plt.xlabel("Customer Segment")
plt.ylabel("Total Quantity")
plt.xticks(rotation=0)
plt.show()
```
