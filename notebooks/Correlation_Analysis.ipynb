## Objective

The objective of correlation analysis is to measure the relationship between numerical variables such as Sales, Profit, Quantity, and Discount.

Understanding these relationships helps identify key business drivers and supports better decision-making.

```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../data/cleaned/superstore_cleaned.csv")
```

### Correlation Matrix

```python
corr = df[["Sales","Profit","Quantity","Discount"]].corr()
corr
```

### HeatMap

```python
plt.figure(figsize=(8,6))
plt.imshow(corr, cmap="coolwarm")
plt.colorbar()
plt.xticks(range(len(corr.columns)), corr.columns, rotation=45)
plt.yticks(range(len(corr.columns)), corr.columns)
plt.title("Correlation Heatmap")
for i in range(len(corr.columns)):
    for j in range(len(corr.columns)):
        plt.text(
            j,
            i,
            round(corr.iloc[i,j],2),
            ha="center",
            va="center",
            color="black"
        )
plt.tight_layout()
plt.show()
```
