## Objective

The objective of this analysis is to understand the distribution of different shipping modes and their impact on sales, profit, and quantity sold.

This analysis helps businesses evaluate customer shipping preferences and optimize logistics.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../data/cleaned/superstore_cleaned.csv")
```

### Number of Orders By ShipMode


```python
ship_count = df["Ship Mode"].value_counts()
print(ship_count)
```

```python
plt.figure(figsize=(8,5))
ship_count.plot(kind="bar", edgecolor="black")
plt.title("Number of Orders by Ship Mode")
plt.xlabel("Ship Mode")
plt.ylabel("Number of Orders")
plt.xticks(rotation=15)
plt.grid(axis="y", linestyle="--", alpha=0.5)
plt.tight_layout()
plt.show()
```

### Sales By ShipMode

```python
sales_ship = (df.groupby("Ship Mode")["Sales"].sum().sort_values(ascending=False))
print(sales_ship)
```

```python
plt.figure(figsize=(8,5))
sales_ship.plot(kind="bar", edgecolor="black")
plt.title("Sales by Ship Mode")
plt.xlabel("Ship Mode")
plt.ylabel("Sales")
plt.xticks(rotation=15)
plt.grid(axis="y", linestyle="--", alpha=0.5)
plt.tight_layout()
plt.show()
```

### Profit By ShipMode

```python
profit_ship = (df.groupby("Ship Mode")["Profit"].sum().sort_values(ascending=False))

print(profit_ship)
```

```python
plt.figure(figsize=(8,5))
profit_ship.plot(kind="bar", edgecolor="black")
plt.title("Profit by Ship Mode")
plt.xlabel("Ship Mode")
plt.ylabel("Profit")
plt.xticks(rotation=15)
plt.grid(axis="y", linestyle="--", alpha=0.5)
plt.tight_layout()
plt.show()
```

### Quantity By ShipMode

```python
quantity_ship = (df.groupby("Ship Mode")["Quantity"].sum().sort_values(ascending=False))

print(quantity_ship)
```

```python
plt.figure(figsize=(8,5))
quantity_ship.plot(kind="bar", edgecolor="black")
plt.title("Quantity Sold by Ship Mode")
plt.xlabel("Ship Mode")
plt.ylabel("Quantity")
plt.xticks(rotation=15)
plt.grid(axis="y", linestyle="--", alpha=0.5)
plt.tight_layout()
plt.show()
```
