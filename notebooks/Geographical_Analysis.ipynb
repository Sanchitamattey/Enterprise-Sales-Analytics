## Objective

The objective of this analysis is to evaluate the geographical distribution of sales and profit across different states, cities, and regions.

This analysis helps identify high-performing markets, loss-making areas, and potential opportunities for business expansion.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
df = pd.read_csv("../data/cleaned/superstore_cleaned.csv")
```

### Top 10 Sales By State

```python
sales_state = (df.groupby("State")["Sales"].sum().sort_values(ascending=False).head(10))

print(sales_state)
```

```python


plt.figure(figsize=(10,6))

sales_state.plot(kind="bar", edgecolor="black")

plt.title("Top 10 States by Sales")
plt.xlabel("State")
plt.ylabel("Sales")

plt.xticks(rotation=45, ha="right")

plt.grid(axis="y", linestyle="--", alpha=0.5)

plt.tight_layout()

plt.show()
```

### Top 10 States By Profit

```python
profit_state = ( df.groupby("State")["Profit"].sum().sort_values(ascending=False).head(10))
print(profit_state)
```

```python
plt.figure(figsize=(10,6))
profit_state.plot(kind="bar", edgecolor="black")
plt.title("Top 10 States by Profit")
plt.xlabel("State")
plt.ylabel("Profit")
plt.xticks(rotation=45, ha="right")
plt.grid(axis="y", linestyle="--", alpha=0.5)
plt.tight_layout()
plt.show()
```
