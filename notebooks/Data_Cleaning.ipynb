 
# Data Cleaning
## Objective
The objective of this notebook is to clean and validate the raw Superstore dataset by identifying missing values, removing duplicate records, verifying data types, validating categorical and numerical columns, and preparing a clean dataset for Exploratory Data Analysis (EDA).

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../data/raw/superstore.csv")
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df = df.drop_duplicates()
```

```python
df.duplicated().sum()
```
