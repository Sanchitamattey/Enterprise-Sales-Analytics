## Objective

The objective of this analysis is to understand the discount patterns offered across different categories and sub-categories and determine whether higher discounts have any impact on profitability.

This analysis helps businesses identify whether discount strategies are increasing sales while maintaining profitability or leading to unnecessary losses.

### Average Discount

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../data/cleaned/superstore_cleaned.csv")
```

```python
average_discount = df["Discount"].mean()
print("Average Discount :", round(average_discount,2))
```

### Discount By Category

```python
discount_category = (df.groupby("Category")["Discount"].mean().sort_values(ascending=False))
print(discount_category)
```

```python
discount_category.plot(kind="bar")
plt.title("Average Discount by Category")
plt.xlabel("Category")
plt.ylabel("Average Discount")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

### Discount By Sub-Category

```python
discount_subcategory = (df.groupby("Sub-Category")["Discount"].mean().sort_values())
print(discount_subcategory)
```

```python
plt.figure(figsize=(12,8))

discount_subcategory.plot(
    kind="barh",
    edgecolor="black"
)

plt.title("Average Discount by Sub-Category", fontsize=14)
plt.xlabel("Average Discount")
plt.ylabel("Sub-Category")

plt.grid(axis="x", linestyle="--", alpha=0.5)

plt.tight_layout()

plt.show()
```

### Discount vs Profit

```python
plt.figure(figsize=(8,6))

plt.scatter(df["Discount"],df["Profit"])

plt.title("Discount vs Profit")

plt.xlabel("Discount")

plt.ylabel("Profit")

plt.show()
```
