# Sales Analysis

Sales by Region

```python
Sales_by_region =df.groupby("Region")["Sales"].sum().sort_values(ascending = False)
print(Sales_by_region)
```

```python
Sales_by_region.plot(kind = "bar",figsize = (8,5))
plt.title("Sales_by_region")
plt.xlabel("Region")
plt.ylabel("Total Sales")
plt.xticks(rotation = 0)
plt.show()
```

```python
Sales_by_category = df.groupby("Category")["Sales"].sum().sort_values(ascending = False)
print(Sales_by_category)
```

```python
Sales_by_category.plot(kind = "bar",figsize =(8,5))
plt.title("Sales_by_category")
plt.xlabel("Category")
plt.ylabel("Total Sales")
plt.xticks(rotation = 0)
plt.show()
```

Sales by Sub-Category

```python
Sales_By_SubCategory = df.groupby("Sub-Category")["Sales"].sum().sort_values(ascending= False)
print(Sales_By_SubCategory)
```

```python
Sales_By_SubCategory.plot(kind = "bar" ,figsize = (10,6))
plt.title("Sales_By_SubCategory")
plt.xlabel("Sales_by_SubCategory")
plt.ylabel("Total_Sales")
plt.xticks(rotation= 90)
plt.show()
```

### Observations

- West region generated the highest sales.
- Technology category contributed the highest revenue.
- Phones were among the highest-selling sub-categories.
